```python
import requests
import aiohttp
import asyncio
import time
from bs4 import BeautifulSoup
from fake_headers import Headers
import pandas as pd
```

```python
global_start = time.time()
```

```python
header = Headers()
```

```python
def get_curr_number():
    currencies = set()
    page = requests.get('https://finance.yahoo.com/crypto', headers=header.generate())
    soup = BeautifulSoup(page.content, 'html5lib')
    curr_number = soup.find('span', attrs={'class': 'Mstart(15px) Fw(500) Fz(s)'})
    if not hasattr(curr_number, 'text'):
        return -1
    curr_number = curr_number.text
    curr_number = int(curr_number.split()[2])
    return curr_number
```

```python
start_time = time.time()
results = []
currencies = set()
i = 0

async def get_html(session, url):
    async with session.get(url, headers=header.generate()) as res:
        print(f'Status: {res.status}, for url={url}')
        html = await res.text()
        return html
    
    
while True:
    time.sleep(3)
    curr_number = get_curr_number()
    if curr_number == -1:
        print("Failed to load curr number")
        print("Waiting for 3 minutes")
        time.sleep(180)
        continue
    elif not (curr_number > len(currencies)):
        break
    
    async def main():
        global results
        async with aiohttp.ClientSession() as session:
            tasks = []
            for offset in range((curr_number // 100) + 1):
                url = f'https://finance.yahoo.com/crypto/?count=100&offset={offset*100}'
                tasks.append(asyncio.ensure_future(get_html(session, url)))

            results = await asyncio.gather(*tasks)
            
            
    await main()
    
    for idx, result in enumerate(results):
        soup = BeautifulSoup(result, 'html.parser')
        curr = soup.find_all('a', attrs={'class': 'Fw(600) C($linkColor)', 'data-test': 'quoteLink'})
        if len(curr) != 0:
            currencies |= set([x.text for x in curr])
    i += 1
    print(f'Iteration: {i}')
    print(f'len={len(currencies)}')
    print('______________')
    print('')
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
flag = False
k = 0
now = int(time.time())
curr_to_iterate = currencies.copy()
curr_to_iterate_2 = currencies.copy()
start_time = time.time()
batch_size = 100

async def download_files(session, url, curr):
    global k, flag
    async with session.get(url, headers=header.generate()) as res:
        if res.status != 200:
            print(f'CSVs-async\\{curr}.csv has failed to save due to href')
            print('Waiting for 1 minute')
            print('______________')
            flag = True
            time.sleep(60)
            return
        text = await res.text()
        with open(f'CSVs-async\\{curr}.csv', 'w') as file_to_save:
            file_to_save.write(text)
        curr_to_iterate_2.remove(curr)
        print(f'CSVs-async\\{curr}.csv has been saved')
        print(f'{k+1} file out of {len(currencies)}')
        print(f'Status code: {res.status}')
        print('______________')
        k += 1

while k < len(currencies):
    start_time = time.time()
    results = []
    

    async def main():
        global results, flag
        async with aiohttp.ClientSession() as session:
            tasks = []
            for curr in curr_to_iterate:
                href = f'https://query1.finance.yahoo.com/v7/finance/download/{curr}?period1={now-60*60*24*365}&period2={now}&interval=1d&events=history&includeAdjustedClose=true'
                tasks.append(asyncio.ensure_future(download_files(session, href, curr)))

            results = await asyncio.gather(*tasks)

    await main()
    curr_to_iterate = curr_to_iterate_2.copy()
print("--- %s seconds ---" % (time.time() - start_time))
```
