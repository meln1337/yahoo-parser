```python
res = requests.get('https://api.coinmarketcap.com/data-api/v3/cryptocurrency/listing?start=1&limit=10000&sortBy=market_cap&sortType=desc&convert=USD&cryptoType=all&tagType=all&audited=false', headers=headers)
data = res.json()
symbols = [x['symbol'] + '-USD' for x in data['data']['cryptoCurrencyList']]
```

```python
symbols
```

```python
currencies = pd.DataFrame({'name': symbols})
currencies.to_csv('currencies.csv', index=False)
```

```python
currencies.value_counts()
```

```python
date_ = date.today()
for idx, symbol in enumerate(symbols):
    df = yf.download(symbol, start=f"{date_.year-1}-{date_.month}-{date_.day}")
    df['Date'] = pd.to_datetime(df.index)
    df.index = range(df.shape[0])
    df['Date'] = df['Date'].dt.strftime('%d-%m-%y')
    df.set_index('Date', inplace=True)
    df.to_csv(f"CSVs\\{symbol}.csv")
    print(f"File {symbol}.csv has been saved")
    print(f"{idx+1} out of {currencies.shape[0]}")
```
