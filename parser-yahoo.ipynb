```python
header = Headers()
```

```python
def get_curr_number():
    currencies = set()
    page = requests.get('https://finance.yahoo.com/crypto', headers=header.generate())
    soup = BeautifulSoup(page.content, 'html5lib')
    curr_number = soup.find('span', attrs={'class': 'Mstart(15px) Fw(500) Fz(s)'})
    if not hasattr(curr_number, 'text'):
        return -1
    curr_number = curr_number.text
    curr_number = int(curr_number.split()[2])
    return curr_number
```

```python
get_curr_number() > len(currencies)
```

```python
start_time = time.time()
currencies = set()
iteration = 0
i = 0
while True:
    time.sleep(3)
    curr_number = get_curr_number()
    if curr_number == -1:
        print("Failed to load curr number")
        print("Waiting for 3 minutes")
        time.sleep(180)
        continue
    elif not (curr_number > len(currencies)):
        break
    print(f'(iteration*100) % curr_number = {(iteration*100) % curr_number}')
    res = requests.get(f'https://finance.yahoo.com/crypto?count=100&offset={(iteration*100) % curr_number}', headers=header.generate())
    if not res.ok:
        print(f'Failed to load: https://finance.yahoo.com/crypto?count=100&offset={(iteration*100) % curr_number}')
        print('Status code:', res.status_code)
        print(f"{iteration+1} page downloading failed")
    else:
        soup = BeautifulSoup(res.content, 'html5lib')
        curr = soup.find_all('a', attrs={'class': 'Fw(600) C($linkColor)', 'data-test': 'quoteLink'})
        if len(curr) != 0:
            currencies |= set([x.text for x in curr])
            print(f"{iteration+1} page")
        else:
            print(f"{iteration+1} failed to load page")
    print(f'Iteration: {iteration+1}, len={len(currencies)}')
    print("___________________________")
    iteration += 1
print("--- %s seconds ---" % (time.time() - start_time))
```
